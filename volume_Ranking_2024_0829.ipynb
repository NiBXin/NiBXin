```python
import re
def remove_brackets_and_content(string):
    pattern = r'\[.*?\]|{.*?}|\(.*?\)'
    new_string = re.sub(pattern, '', string)
    return new_string
```

```python
da = {'company name':[1],'code':[1],'share':[1]}
final_data = pd.DataFrame(da)
# print(final_data)
error_list = []
for code in data['code'].tolist():
    error_time = 0
    wait = 0
    service = Service("C:\Program Files (x86)\Google\Chrome\Application\chromedriver")  
    browser = webdriver.Chrome(service=service, options=chrome_options)
    url_code = "https://www.marketindex.com.au/asx/" + code.lower() + "?src=search-all"
    browser.get(url_code)
    while wait == 0:
        time.sleep(1)
        wait = 1
        try:
            soup = BeautifulSoup(browser.page_source, 'lxml')
            company = soup.find('h1')
            company_name = company.text.strip()
        except  Exception as error:
            print("An exception occurred:", error)
            wait = 0
    company_name = remove_brackets_and_content(company_name).strip()
    print(company_name)
    print(code)
    share_list = []
    wait = 0
    while wait == 0 and error_time <= 15:
        time.sleep(1)
        wait = 1
        try:
            top20_table = soup.find('div',id = 'top-shareholders-root').table.tbody
            print(111)
#             while len(top20_table.find_all('tr')) == 0:
#                 time.sleep(1)
            for name in top20_table.find_all('tr'):
                share_list.append(name.td.text.strip())
        except  Exception as error:
            error_time = error_time + 1
            print("An exception occurred:", error)
            wait = 0
    if error_time >= 15:
        error_list.append(code)
    print(share_list)
    share_data = {'share':share_list}
    temp_data = pd.DataFrame(share_data)
    temp_data.insert(0, 'code', code)
    temp_data.insert(0, 'company name', company_name)
#     print(temp_data)
    final_data = pd.concat([final_data,temp_data], ignore_index=True)
    browser.close()
```
