```python
def llama_question_answer(content):
    question1 = ". Question: What or who act as the broker or who is broker? if not have or you can't make sure, just answer 'No.'.Don't explain."
    question2 = ". Question: What or who act as the broker? Just give me the name.In the format of A,B(if it has more than one name). I don't need other responses.Don't explain."
    
    prompt_check_broker = content + question1
    prompt_get_broker = content + question2
    url = 'http://localhost:11434/api/generate'
    headers = {"Content-Type":"application/json"}
    data = {
        "model" : "llama3.1",
        "prompt":  prompt_check_broker,
        "stream": False,
        "temperature":0,
        'top_k':1
    }
    response1 = requests.post(url,  data = json.dumps(data))
    
    answer1 = json.loads(response1.text)['response']
    print(answer1)
    
    answer2 = ''
    if 'No.'not in answer1:
        url = 'http://localhost:11434/api/generate'
        headers = {"Content-Type":"application/json"}
        data = {
            "model" : "llama3.1",
            "prompt":  prompt_get_broker,
            "stream": False,
            "temperature":0
        }
        response2 = requests.post(url,  data = json.dumps(data))
        answer2 = json.loads(response2.text)['response']
    return answer2
```

```python
def get_brokers(content):
    answer_list = []
    for attempt in range(0,5):
        answer = llama_question_answer(content)
        answer_list.append(answer)
    f_answer = max(answer_list, key = answer_list.count)
    return f_answer
```

```python
def run_threads(cap_raise_info):
    threads = []
    n_threads = 10
    interval = len(cap_raise_info)//n_threads
    for thread_index in range(n_threads):
        start = interval*thread_index
        end = interval*(thread_index + 1)
        if thread_index == n_threads - 1:
            end = len(cap_raise_info)
        individual_thread = threading.Thread(target=find_manager_brokers, args=(cap_raise_info[start:end],))
        threads.append(individual_thread)
        individual_thread.start()
#         time.sleep(random.randint(1, 5))
        
    for individual_thread in threads:
        individual_thread.join()
```
