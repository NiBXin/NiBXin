```python
def write_excel_xlsx1(path, sheet_name, value):
    index = len(value)
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = sheet_name
    for index, i in enumerate(value):
        sheet.cell(row=index+1, column=1, value=str(i))
        print(i)
        print(value[i][0][0])
        for j in range(0,len(value[i])):
            sheet.cell(row=index+1, column=2+j, value=str(value[i][j][0]))
    workbook.save(path)
    print("xlsx write successfullyÔºÅ")
```

```python
def write_excel_xlsx2(path, sheet_name, value):
    index = len(value)
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = sheet_name
    for index, i in enumerate(value):
        sheet.cell(row=index+1, column=1, value=str(value[index][0]))
        sheet.cell(row=index+1, column=2, value=str(value[index][1]))
        print(i)
        print(value[index][0][0])
        for j in range(0,len(value[index][2])):
            sheet.cell(row=index+1, column=3+j, value=str(value[index][2][j]))
    workbook.save(path)
    print("xlsx write successfullyÔºÅ")
```
